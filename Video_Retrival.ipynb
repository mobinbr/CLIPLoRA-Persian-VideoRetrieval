## 2. Download the dataset and check it out

```python
import pandas as pd
from datasets import load_dataset

ds = load_dataset("lmms-lab/YouCook2")
df = pd.DataFrame(ds['val'])
df[['start', 'end']] = pd.DataFrame(df['segment'].tolist(), index=df.index)
rm_cols = ['id','video_url', 'recipe_type', 'segment', 'video_path']
df.drop(columns=rm_cols, inplace=True)
df['translated_sentence'] = pd.Series('', dtype='str')
gdf = df.groupby('youtube_id')
```

## 3. Translate the annotations

```python
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='en', target='fa')

for gn, gd in gdf:
    joined_sentences = "\n".join(gd['sentence'])
    translated_list = translator.translate(joined_sentences).split("\n")
    for idx, val in zip(gd.index, translated_list):
      df.at[idx, 'translated_sentence'] = val
```

## 4. Make the **embeddings** ready and check their precsion

```python
def preprocess_persian_text(sentence):
    # Define a simple list of Persian stopwords (if available)
    persian_stop_words = {'و', 'به', 'از', 'که', 'این', 'را', 'با', 'در', 'برای', 'تا'}

    # Tokenize by splitting on spaces
    tokens = sentence.split()

    # Remove stopwords
    filtered_tokens = [word for word in tokens if word not in persian_stop_words]

    return " ".join(filtered_tokens)

df['preprocessed_translated_sentence'] = df['translated_sentence'].apply(preprocess_persian_text)
```

```python
from sentence_transformers import SentenceTransformer
from scipy.spatial.distance import cdist
import statistics

# Load the embedding model
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-mpnet-base-v2')

# Extract English and Persian sentences without identifiers
english_sentences = df['sentence']
clean_persian_sentences = df['translated_sentence']

# Encode English and Persian sentences in batches
english_embeddings = model.encode(english_sentences, batch_size=64)
persian_embeddings = model.encode(clean_persian_sentences, batch_size=64)

# Compute cosine similarity (1 - cosine distance)
similarity_scores = 1 - cdist(english_embeddings, persian_embeddings, metric='cosine').diagonal()
df['similarity_score'] = similarity_scores
avg_similarity = statistics.mean(similarity_scores)

# Print results
print("Similarity Scores:", similarity_scores[:5])
print("Average Similarity:", avg_similarity)
```

## 5. Prepare the video embeddings (step by step)

### 5.1 Download Videos for Segments

```python
import os
import subprocess
import pdb

def download_video(youtube_id, save_path="videos/"):
    # pdb.set_trace()
    os.makedirs(save_path, exist_ok=True)
    url = f"https://www.youtube.com/watch?v={youtube_id}"
    output_template = os.path.join(save_path, f"{youtube_id}.mp4")
    try:
        subprocess.run(["yt-dlp", "-f", "best[ext=mp4]", "-o", output_template, url], check=True)
        return output_template
    except subprocess.CalledProcessError as e:
        print(f"Error downloading {youtube_id}: {e}")
        return None
```

### 5.2 Extract Frames for Segments

```python
import cv2
import os

def extract_frames(video_path, gd, frame_rate=1, output_base_folder="frames/"):
    segment_folder = os.path.join(output_base_folder, os.path.basename(video_path).split('.')[0])
    os.makedirs(segment_folder, exist_ok=True)

    cap = cv2.VideoCapture(video_path)

    for index, row in gd.iterrows():
        extract_segment_frames(segment_folder, cap, row['start'], row['end'], frame_rate)

    cap.release()
    print(f"Segments saved in: {segment_folder}")

def extract_segment_frames(segment_folder, cap, start, end, frame_rate):
    segment_subfolder = f"{start}_{end}"
    output_folder = os.path.join(segment_folder, segment_subfolder)
    os.makedirs(output_folder, exist_ok=True)

    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = start * fps
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_count)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        time = frame_count / fps
        if start <= time <= end:
            if frame_count % max(1, int(fps / frame_rate)) == 0:  # Save frames at the specified frame rate
                frame_name = os.path.join(output_folder, f"frame_{frame_count}.jpg")
                cv2.imwrite(frame_name, frame)

        frame_count += 1
        if time > end:
            break

    print(f"Frames saved in: {output_folder}")
```

### 5.4 Pipeline Integration
Combine all steps:
1. Download the video using `download_video`.
2. Extract frames using `extract_frames`.
3. Compute embeddings using `get_frame_embeddings`.

```python
import os

print(f"Total videos to process: {len(gdf)}")

# Configuration
checkpoint_file = "processed_segments.txt"

# Load previously processed video segments
if os.path.exists(checkpoint_file):
    with open(checkpoint_file, "r") as f:
        processed_segments = set(f.read().splitlines())
else:
    processed_segments = set()

# Process videos in batches
for gn, gd in gdf:
    print(f"Processing videos {gn}")
    video_file = download_video(gn)

    if gn in processed_segments:
            print(f"Skipping already processed segment: {gn}")
            continue  # Skip already processed segments

    if video_file:  # Ensure video was successfully downloaded
        extract_frames(video_file, gd, frame_rate=0.5)

    # Save progress to checkpoint file
    with open(checkpoint_file, "a") as f:
        f.write(f"{gn}\n")

    print(f"Finished vide {gn}. Progress saved.")

print("All segments processed successfully!")
```

```python
import torch
from PIL import Image
import pandas as pd
from torch.utils.data import Dataset

class ImageTextDataset(Dataset):
    def __init__(self, data, processor):
        self.data = data # CSV with "image_path" and "caption" columns
        self.processor = processor

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.data[idx]["image"]).convert("RGB")
        except Exception as e:
            print(f"Error loading image {self.data.iloc[idx]['image']}: {e}")
            return None  # Skip problematic images

        text = self.data[idx]["text"]

        return {"image": image, "text": text}

def collate_fn(batch):
    images, texts = [], []

    for item in batch:
        if item and item["image"] is not None and item["text"] is not None:
            images.append(item["image"])
            texts.append(item["text"])

    if len(images) == 0:
        return None

    # Process images & texts with padding
    inputs = processor(text=texts, images=images, return_tensors="pt", padding=True)

    return inputs
```

```python
from peft import get_peft_model, LoraConfig, TaskType
from transformers import (
    VisionTextDualEncoderConfig,
    VisionTextDualEncoderModel,
    CLIPVisionModel,
    RobertaModel,
    CLIPVisionConfig,
    RobertaConfig
)
import torch.nn.init as init
from torch import nn
import copy

r=16
lora_alpha = 16
lora_dropout = 0.1

def freeze_and_unfreeze_layers(model):
    # Freeze all parameters first
    for param in model.parameters():
        param.requires_grad = False

    print("freezing completed.")
    return model

def apply_lora(text_encoder):


    config = LoraConfig(
        r=r,
        lora_alpha=lora_alpha,
         target_modules=[
        "attention.self.query",
        "attention.self.value",
        "q_proj", "v_proj"
        ],
        lora_dropout=lora_dropout,
        bias="none",
    )
    lora_model = get_peft_model(text_encoder, config)
    # model = get_peft_model(model, lora_config, target_modules=["pooler", "encoder.layer.10", "encoder.layer.11"])

    return lora_model


def model_clipfa(
    text_model_name="SajjadAyoubi/clip-fa-text",
    vision_model_name="SajjadAyoubi/clip-fa-vision",
    feature_extractor_name = "SajjadAyoubi/clip-fa-vision"
):
    vision_config = CLIPVisionConfig.from_pretrained(vision_model_name)
    text_config = RobertaConfig.from_pretrained(text_model_name)

    dual_encoder_config = VisionTextDualEncoderConfig.from_vision_text_configs(
        vision_config=vision_config,
        text_config=text_config,
        projection_dim=768
    )

    assert vision_config.hidden_size == text_config.hidden_size

    print("vision_config.hidden_size: " ,vision_config.hidden_size )
    print("text_config.hidden_size: " ,text_config.hidden_size )


    model = VisionTextDualEncoderModel(config=dual_encoder_config)

    vision_encoder = CLIPVisionModel.from_pretrained(vision_model_name)
    text_encoder = RobertaModel.from_pretrained(text_model_name)

    vision_encoder = freeze_and_unfreeze_layers(vision_encoder)
    text_encoder = freeze_and_unfreeze_layers(text_encoder)

    model.vision_model = vision_encoder
    model.text_model = text_encoder

    print("visual_projection: " , model.visual_projection.weight.shape)
    model.visual_projection = nn.Identity()
    model.text_projection = nn.Identity()
    # base_model = model
    # model.save_pretrained(output_dir)

    model = apply_lora(model)


    tokenizer = AutoTokenizer.from_pretrained(text_model_name)
    vision_processor = CLIPFeatureExtractor.from_pretrained(feature_extractor_name)

    return model, tokenizer, vision_processor
```

```python
from PIL import Image
import glob
import os
import pandas as pd
from torch.utils.data import DataLoader

# Load video frames
frame_folder = "/content/frames/"

def create_dataloader(csv_file, frame_folder, batch_size=8):
    df = pd.read_csv(csv_file)

    samples = []
    for idx, row in df.iterrows():
        persian_text = row['preprocessed_translated_sentence']
        video_name = row['youtube_id']
        segment_time = f"{row['start']}_{row['end']}"

        video_folder = os.path.join(frame_folder, video_name, segment_time)
        if os.path.exists(video_folder):
            frame_files = sorted(glob.glob(os.path.join(video_folder, "*.jpg")))
            if frame_files:
                samples.append((persian_text, frame_files[len(frame_files)//2]))

    dataset = [{"text": text, "image": img_path} for text, img_path in samples]
    dataset = ImageTextDataset(dataset, processor)
    dataLoader = DataLoader(dataset, batch_size=8, shuffle=True, collate_fn=collate_fn)
    optimizer = torch.optim.AdamW(model.parameters(), lr=5e-6)

    return dataLoader, optimizer
```

```python
csv_file = "/content/translation_quality_updated.csv"
batch_size = 8
dataloader, optimizer = create_dataloader(csv_file, frame_folder, batch_size=batch_size)
```

```python
from torch.optim import AdamW
from torch.optim.lr_scheduler import StepLR
import torch.nn.functional as F
from transformers import get_linear_schedule_with_warmup
from torchvision import transforms

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Optimizer with weight decay
num_epochs = 3
model.train()
for epoch in range(num_epochs):
    total_loss = 0

    for batch in dataloader:
        if batch is None:
            continue

        optimizer.zero_grad()

        inputs = {key: val.to(device) for key, val in batch.items()}

        outputs = model(**inputs)

        image_embeds = outputs.image_embeds
        text_embeds = outputs.text_embeds

        # Ensure embeddings match in size
        batch_size = min(image_embeds.shape[0], text_embeds.shape[0])
        image_embeds = image_embeds[:batch_size]
        text_embeds = text_embeds[:batch_size]

        # Normalize embeddings
        image_embeds = F.normalize(image_embeds, p=2, dim=-1)
        text_embeds = F.normalize(text_embeds, p=2, dim=-1)

        # Compute cosine similarity matrix
        logits_per_image = image_embeds @ text_embeds.T
        logits_per_text = logits_per_image.T

        # Ground truth labels
        labels = torch.arange(batch_size, device=device)

        # Compute loss
        loss = (F.cross_entropy(logits_per_image, labels) + F.cross_entropy(logits_per_text, labels)) / 2

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {total_loss/len(dataloader)}")
```

```python
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import pandas as pd
import glob

# Load the fine-tuned model and processor
model_path = "fine_tuned_clip_lora"
model = CLIPModel.from_pretrained(model_path).to("cuda" if torch.cuda.is_available() else "cpu")
processor = CLIPProcessor.from_pretrained(model_path)

# Load dataset metadata (assuming a CSV file with video IDs and image paths)
csv_file = "translation_quality_updated.csv"
df = pd.read_csv(csv_file)

# Prepare image paths and video IDs
image_paths = []
video_ids = []
for idx, row in df.iterrows():
    frame_path = glob.glob(f"/content/frames/{row['youtube_id']}/{row['start']}_{row['end']}/*.jpg")
    if frame_path:
        image_paths.append(frame_path[0])
        video_ids.append(row)

# Function to retrieve top 5 most related video IDs
def retrieve_top_videos(query_text, image_paths, video_ids, top_k=5):
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Preprocess input text
    inputs_text = processor(text=[query_text], return_tensors="pt").to(device)

    # Encode text embedding
    with torch.no_grad():
        text_embeds = model.get_text_features(**inputs_text)

    # Prepare image embeddings
    image_embeds = []
    for img_path in image_paths:
        image = Image.open(img_path).convert("RGB")
        inputs_image = processor(images=image, return_tensors="pt").to(device)

        with torch.no_grad():
            image_embed = model.get_image_features(**inputs_image)

        image_embeds.append(image_embed)

    # Convert image embeddings list to tensor
    image_embeds = torch.cat(image_embeds, dim=0)

    # Compute similarity (cosine similarity)
    similarities = torch.cosine_similarity(text_embeds, image_embeds)

    # Get the top k related video indices
    top_indices = torch.argsort(similarities, descending=True)[:top_k]

    # Retrieve the corresponding video IDs
    top_video_ids = [video_ids[i] for i in top_indices.cpu().numpy()]
    top_similarities = [similarities[i].item() for i in top_indices.cpu().numpy()]

    return list(zip(top_video_ids, top_similarities))

query = "سیب زمینی"
top_videos = retrieve_top_videos(query, image_paths, video_ids)

# Print top 5 related videos
print("Top 5 Related Videos:")
for vid, score in top_videos:
    print(f"Video ID: {vid['youtube_id']}, segment: {row['start']}_{row['end']}, Similarity: {score:.4f}")
```
